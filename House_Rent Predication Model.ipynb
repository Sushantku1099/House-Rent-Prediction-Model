**House_Rent Predication Model using Simple Linear Regression** ----> **Rent** is A Dependent Variable, **Size** is a Independent Variable.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/House-Size_Rent.csv')
Dataset['Rent (?/month)'] = Dataset['Rent (?/month)'].str.replace(',', '').astype(float)
Dataset
```

```python
x = Dataset.iloc[:,:-1].values
x
```

```python
y = Dataset.iloc[:,1].values
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
Lr = LinearRegression()
Lr.fit(x_train,y_train)
```

```python
Lr.predict(x_test)
```

```python
Lr.predict([[1450]])
```

```python
# y = a + bx
Lr.coef_  #b
```

```python
Lr.intercept_ #a
```

```python
-831.9575471698099 + 15.54245283 * 1450
```

```python
NewDataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/House-Size.csv')
NewDataset
```

```python
S = Lr.predict(NewDataset)
S
```

```python
df = pd.DataFrame(NewDataset) # Convert Array To Normal Form
df ["Rent (?/month)"] = S
df
```

```python
df.to_csv('Predicated House_Rent.csv')
df.to_excel('Predicated House_Rent.xlsx')
```
